```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium

# 
data = pd.read_csv('202409-citibike-tripdata/202409-citibike-tripdata_1.csv')

# 
data['started_at'] = pd.to_datetime(data['started_at'])
data['ended_at'] = pd.to_datetime(data['ended_at'])

# 
data['ride_duration'] = (data['ended_at'] - data['started_at']).dt.total_seconds() / 60

# 
plt.figure(figsize=(10, 6))
plt.hist(data['ride_duration'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Ride Duration (in minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.xlim(0, 60)
plt.show()

# 
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='rideable_type', palette='viridis')
plt.title('Number of Rides by Rideable Type')
plt.xlabel('Rideable Type')
plt.ylabel('Count')
plt.show()

# 
member_counts = data['member_casual'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(member_counts, labels=member_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Proportion of Member vs Casual Rides')
plt.show()

# 
start_station_counts = data['start_station_name'].value_counts().head(10)

# 
plt.figure(figsize=(12, 8))
sns.barplot(y=start_station_counts.index, x=start_station_counts.values, palette='coolwarm')
plt.title('Top 10 Start Stations by Number of Rides')
plt.xlabel('Number of Rides')
plt.ylabel('Start Station')
plt.show()

# JC
stations = data[['start_station_name', 'start_lat', 'start_lng']].drop_duplicates()
m = folium.Map(location=[stations['start_lat'].mean(), stations['start_lng'].mean()], zoom_start=13)

# 
for _, row in stations.iterrows():
    folium.Marker(
        location=[row['start_lat'], row['start_lng']],
        popup=row['start_station_name'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

# 
m.save('jc_stations_map.html')
```

```python
# 
data['hour'] = data['started_at'].dt.hour

# 24
hourly_usage = data.groupby(['start_station_name', 'hour']).size().reset_index(name='count')

# 24
plt.figure(figsize=(14, 10))
sns.lineplot(data=hourly_usage, x='hour', y='count', hue='start_station_name', palette='tab10')
plt.title('Hourly Usage Frequency of Each Station')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Rides')
plt.legend(title='Start Station', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 24
hourly_usage = data.groupby(['start_station_name', 'hour']).size().unstack(fill_value=0)

# 
plt.figure(figsize=(14, 12))
sns.heatmap(hourly_usage, cmap='YlGnBu', linewidths=0.5)
plt.title('Hourly Usage Frequency Heatmap of Each Station')
plt.xlabel('Hour of the Day')
plt.ylabel('Start Station')
plt.xticks(ticks=range(24), labels=range(24))
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# 
station_coords = data[['start_station_name', 'start_lat', 'start_lng']].drop_duplicates()

# K-Means
num_clusters = 5  # 
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
station_coords['cluster'] = kmeans.fit_predict(station_coords[['start_lat', 'start_lng']])

# 
m = folium.Map(location=[station_coords['start_lat'].mean(), station_coords['start_lng'].mean()], zoom_start=13)

# 
colors = ['red', 'blue', 'green', 'purple', 'orange']

# 
for _, row in station_coords.iterrows():
    folium.Marker(
        location=[row['start_lat'], row['start_lng']],
        popup=f"{row['start_station_name']} - Cluster {row['cluster']}",
        icon=folium.Icon(color=colors[row['cluster']], icon='info-sign')
    ).add_to(m)

# 
m.save('jc_stations_clustered_map.html')
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import folium
from matplotlib import cm, colors

# 
station_coords = data[['start_station_name', 'start_lat', 'start_lng']].drop_duplicates()

# K-Means
num_clusters = 20  # 
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
station_coords['cluster'] = kmeans.fit_predict(station_coords[['start_lat', 'start_lng']])

# 
m = folium.Map(location=[station_coords['start_lat'].mean(), station_coords['start_lng'].mean()], zoom_start=13)

# 
cmap = cm.get_cmap('tab10', num_clusters)

# 
for _, row in station_coords.iterrows():
    cluster_color = colors.rgb2hex(cmap(row['cluster'] / num_clusters))
    folium.Marker(
        location=[row['start_lat'], row['start_lng']],
        popup=f"{row['start_station_name']} - Cluster {row['cluster']}",
        icon=folium.Icon(color='blue', icon='info-sign', icon_color=cluster_color)
    ).add_to(m)

# 
m.save('stations_clustered_map.html')
```
